```python
import pandas as pd

rf_base_results = pd.read_csv('.\\random_forest_base_model_results.csv')
rf_tuned_results = pd.read_csv('.\\random_forest_evaluation_results_after_tuning.csv')
lr_base_results = pd.read_csv('.\\logistic_model_evaluation_results.csv')
lr_tuned_results = pd.read_csv('.\\logistic_model_evaluation_results_after_tuning.csv')
xgb_base_results = pd.read_csv('.\\xgboost_model_evaluation_results.csv')
xgb_tuned_results = pd.read_csv('.\\xgboost_model_evaluation_results_after_tuning.csv')
lgb_base_results = pd.read_csv('.\\lightgbm_model_evaluation_results.csv')
lgb_tuned_results = pd.read_csv('.\\lightgbm_model_evaluation_results_after_tuning.csv')
```

```python
print(rf_base_results.columns)
print(rf_tuned_results.columns)
print(lr_base_results.columns)
print(lr_tuned_results.columns)
print(xgb_base_results.columns)
print(xgb_tuned_results.columns)
print(lgb_base_results.columns)
print(lgb_tuned_results.columns)
```

```python
for df in [rf_base_results, rf_tuned_results, lr_base_results, lr_tuned_results, 
           xgb_base_results, xgb_tuned_results, lgb_base_results, lgb_tuned_results]:
    df.rename(columns={'Unnamed: 0': 'Skill'}, inplace=True)

# Standardize column names for consistency
for df in [rf_tuned_results, lr_tuned_results, xgb_tuned_results, lgb_tuned_results]:
    df.rename(columns={'F1-score': 'F1-Score'}, inplace=True)  # Rename 'F1-score' to 'F1-Score'

# Strip any leading or trailing spaces from column names
for df in [rf_base_results, rf_tuned_results, lr_base_results, lr_tuned_results, 
           xgb_base_results, xgb_tuned_results, lgb_base_results, lgb_tuned_results]:
    df.columns = df.columns.str.strip()  # Strip whitespace from column names

# Define the function to prepare metrics for comparison
def add_metrics(df, model_name, tuned=False):
    suffix = '_tuned' if tuned else '_base'
    df.columns = ['Skill', 
                  f'{model_name}_Accuracy{suffix}', 
                  f'{model_name}_Precision{suffix}', 
                  f'{model_name}_Recall{suffix}', 
                  f'{model_name}_F1-Score{suffix}']
    return df

# Initialize metrics comparison DataFrame with the first model's metrics
metrics_comparison = add_metrics(rf_base_results, 'RF', tuned=False)

# Add the tuned metrics for RF
metrics_comparison = pd.merge(metrics_comparison, add_metrics(rf_tuned_results, 'RF', tuned=True), on='Skill', how='outer')

# Add metrics from other models
metrics_comparison = pd.merge(metrics_comparison, add_metrics(lr_base_results, 'LR', tuned=False), on='Skill', how='outer')
metrics_comparison = pd.merge(metrics_comparison, add_metrics(lr_tuned_results, 'LR', tuned=True), on='Skill', how='outer')
metrics_comparison = pd.merge(metrics_comparison, add_metrics(xgb_base_results, 'XGB', tuned=False), on='Skill', how='outer')
metrics_comparison = pd.merge(metrics_comparison, add_metrics(xgb_tuned_results, 'XGB', tuned=True), on='Skill', how='outer')
metrics_comparison = pd.merge(metrics_comparison, add_metrics(lgb_base_results, 'LGB', tuned=False), on='Skill', how='outer')
metrics_comparison = pd.merge(metrics_comparison, add_metrics(lgb_tuned_results, 'LGB', tuned=True), on='Skill', how='outer')

metrics_comparison
```

```python
# Calculate average metrics for each model
avg_metrics = {}

# List of model names and their corresponding metrics
models = ['RF', 'LR', 'XGB', 'LGB']
for model in models:
    avg_metrics[f'{model} Base'] = {
        'Accuracy': metrics_comparison[f'{model}_Accuracy_base'].mean(),
        'Precision': metrics_comparison[f'{model}_Precision_base'].mean(),
        'Recall': metrics_comparison[f'{model}_Recall_base'].mean(),
        'F1-Score': metrics_comparison[f'{model}_F1-Score_base'].mean(),
    }
    avg_metrics[f'{model} Tuned'] = {
        'Accuracy': metrics_comparison[f'{model}_Accuracy_tuned'].mean(),
        'Precision': metrics_comparison[f'{model}_Precision_tuned'].mean(),
        'Recall': metrics_comparison[f'{model}_Recall_tuned'].mean(),
        'F1-Score': metrics_comparison[f'{model}_F1-Score_tuned'].mean(),
    }

# Convert average metrics to a DataFrame for better visualization
avg_metrics_df = pd.DataFrame(avg_metrics).T

# Display the average metrics
print(avg_metrics_df)

# Determine the best model based on average metrics
best_models = avg_metrics_df.idxmax()  # Find the best model for each metric
print("\nBest Model for Each Metric:")
print(best_models)
```

```python
# Create a DataFrame for the best models
best_models_df = pd.DataFrame(best_models, columns=['Best Model'])

# Formatting the output for saving
output_df = pd.concat([avg_metrics_df, pd.DataFrame(index=['']), best_models_df.T], axis=0)
output_df.index = output_df.index.astype(str)  # Ensure the index is string type for better formatting
output_df.index = output_df.index.str.replace("Best Model", "Best Model for Each Metric", regex=False)

# Save to a CSV file
output_df.to_csv('best_model_metrics_summary.csv')

# Display the combined DataFrame
print(output_df)
```
